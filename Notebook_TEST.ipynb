# Exemplo 1

Este é um plot de uma função $seno$


```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size': 17})

Fs = 8000
f = 5
sample = 8000
x = np.arange(sample)
y = np.sin(2 * np.pi * f * x / Fs)
plt.plot(x, y)
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.show()
```

## Simple plotting exemple 2

this is a plot of $sin$ function

$$
f(x) = \frac{\sin(x)}{x}
$$

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size': 17})
import numpy as np

import matplotlib.pyplot as plt
pi= np.pi
x = np.linspace(-4*pi, 4*pi, 1000)
plt.xticks([-4*pi, -2*pi, 0 , 2*pi, 4*pi],
          ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'])
plt.yticks([0, 1], ['$0$', '$1$'])
#Título do gráfico:
plt.title('The $sinc$ function')

#colocando a linha da origem:
plt.axhline(0, color='black', lw=1)

#se as informações do eixo y não aparecerem usar esta função:
#plt.ylim(-0.5, 1.5)
plt.plot(x, np.sin(x)/x, label=r'$f(x) = \frac{\sin(x)}{x}$')
#para mostrar a legenda:
# loc='best' coloca a legenda no melhor lugar do gráfico sem ficar na frente de 
# nenhuma parte do grápico
plt.legend(loc ='best', fontsize = 20)
plt.show()
```
